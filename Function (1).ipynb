# Functions

1. What is the difference between a function and a method in Python?
- Function

  A block of code that performs a specific task.

  Defined using def.

  Can exist independently (outside any class).

  Called using its name: my_function().



-  Method

   A function that belongs to a class or object.

   Always defined inside a class.

   Called using an object: object.method().

    It automatically receives the first argument self (for instance methods).
2.  Explain the concept of function arguments and parameters in Python.
- Parameters

  These are the variable names written in the function definition.

  They act like placeholders for the values the function will receive.

-  Arguments

   These are the actual values you pass to the function when calling it.

    They fill the placeholders (parameters).
3. What are the different ways to define and call a function in Python?
- Normal Function (Most Common Way)

  Define

      def greet():

        greet()

-  Function with Parameters

   Define

       def add(a, b):

   A. Using *args (multiple positional arguments)

       def total(*numbers):

            total(1, 2, 3, 4)

       def info(**details):

    Call:

        info(name="Ashish", age=20)

-  Lambda Function (Anonymous Function)
    Define

          square = lambda x: x * x

    Call

    Define

        obj = MathTool()

    Define

4. What is the purpose of the `return` statement in a Python function?    
- Sends a Value Back to the Caller

  Whatever you put after return goes back to where the function was called.

               def add(a, b):
                 return a + b

                 x = add(5, 3)   # x now becomes 8

-  Ends the Function Immediately

   Code after return will not run.

        def test():
          return 10
          print("This will not execute")

- Allows the Function to Produce Output

   Without return, a function gives back None.

             def greet():
               print("Hi!")

             result = greet()
              print(result)   # Output: None

-  Can Return Multiple Values (as a tuple)

             def calc(a, b):
               return a + b, a * b

    s, m = calc(3, 4)   # s = 7, m = 12

- Helps in Further Calculations

    Returned values can be used in expressions.

     total = add(2, 3) * 5

 6.  Explain the concept of generators in Python and how they are defined.
 - What is a Generator?

    A special type of iterator.

    Generates values one at a time instead of storing them in memory.

    More memory-efficient than lists.

    Perfect for large data processing.

- How Generators Work

    They use the yield keyword instead of return.

    Every time yield is hit, the function pauses and returns a value.

    When called again, it continues from where it paused.

- How to Define a Generator
   A. Generator Function (uses yield)

         def count_up_to(n):
           i = 1
           while i <= n:
            yield i
              i += 1

     Calling It

          gen = count_up_to(5)

           print(next(gen))   # 1
           print(next(gen))   # 2


    It behaves just like an iterator.

  B. Generator Expression

      Similar to list comprehension but uses ( ) instead of [ ].

            gen = (x * x for x in range(5))

     Call

           next(gen)   # gives 0, then 1, then 4...

9.  Explain the purpose and usage of the `map()` function in Python.
- The map() function is used to apply a function to every item
   in an iterable (like a list, tuple, etc.) and return a map object (iterator) with the results.

  Syntax
             map(function, iterable)

  How It Works

- Takes a function

- Takes an iterable

- Applies the function to each element

- Returns results one by one (lazy evaluation)

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- map()

   Purpose:

    Applies a function to every element in an iterable.

   What it does:

   Transforms data.

   Example:

           map(lambda x: x * 2, [1, 2, 3])  →  [2, 4, 6]

- filter()

    Purpose:

    Keeps only those elements for which the function returns True.

    What it does:

    Filters data.

    Example:

            filter(lambda x: x % 2 == 0, [1, 2, 3, 4])  →  [2, 4]

- reduce()

    Purpose:

    Combines all elements into a single output value.
     (Needs to be imported from functools.)

     What it does:

    Aggregates or reduces data.

    Example:

            reduce(lambda a, b: a + b, [1, 2, 3, 4])  →  10

```python
# Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total


nums = [10, 15, 22, 33, 40]

result = sum_of_even(nums)
print(result)
```

```python
# Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
num=int(input("enter nuber: "))
print(is_prime(num))
```

```python
#  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_value = self.a + self.b
            self.a = self.b
            self.b = next_value
            self.count += 1
            return next_value




n = int(input("Enter number of Fibonacci terms: "))

fib = FibonacciIterator(n)

for num in fib:
    print(num)
```

```python
# Write a generator function in Python that yields the powers of 2 up to a given exponent
def power_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

exp = int(input("Enter exponent: "))

for value in power_of_two(exp):
    print(value)
```

```python
#  Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()   # strip() → newline remove

for line in read_file_lines("data.txt"):
    print(line)
```

```python
# Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(3, 20), (1, 5), (4, 15), (2, 8)]

sorted_list = sorted(data, key=lambda x: x[1])

print(sorted_list)
```

```python
# Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius = [0, 20, 37, 100]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print(fahrenheit)
```

```python
# Create a Python program that uses `filter()` to remove all the vowels from a given string.
text = "Hello World"

vowels = "aeiouAEIOU"
result = ''.join(filter(lambda char: char not in vowels, text))

print(result)
```

```python
""" Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

order no.       Book Title & Author                     Quantity         per Item
34587           Learning python,Mark Lutz                  4                40.95
98762           Programming python,Mark Lutz               5                56.80
77226           Head First Python,Paul Barry               3                32.95
88112           Einfuhrung in python3,Bernd Klein          3                24.99

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map."""


# List of orders: [order_no, book title & author, quantity, price per item]
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bernd Klein", 3, 24.99]
]

# Using map and lambda to calculate total price
totals = list(map(lambda order: (
    order[0],  # order number
    order[2] * order[3] if order[2] * order[3] >= 100 else order[2] * order[3] + 10
), orders))

print(totals)
```
